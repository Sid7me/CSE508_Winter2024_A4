```python
import pandas as pd
from tabulate import tabulate

file_path = 'archive\Reviews.csv'

df = pd.read_csv(file_path)

data = df.values.tolist()[:30] 

headers = df.columns.tolist()

print(tabulate(data, headers=headers, tablefmt='fancy_grid'))
```

```python
from bs4 import BeautifulSoup
from unidecode import unidecode
import re
import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

def remove_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')
    return soup.get_text()

def remove_accented_chars(text):
    return unidecode(text)

def remove_special_characters(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)

def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    return ' '.join([lemmatizer.lemmatize(word, wordnet.VERB) for word in text.split()])

def normalize_text(text):
    return text.lower().strip()

def preprocess_review_text(review_text):
    if isinstance(review_text, str):  
        review_text = remove_html_tags(review_text)
        review_text = remove_accented_chars(review_text)
        review_text = remove_special_characters(review_text)
        review_text = lemmatize_text(review_text)
        review_text = normalize_text(review_text)
        return review_text
    else:  
        return ''
```

```python

df['Text'] = df['Text'].apply(preprocess_review_text)
df['Summary'] = df['Summary'].apply(preprocess_review_text)

data = df.values.tolist()[:30] 

headers = df.columns.tolist()

print(tabulate(data, headers=headers, tablefmt='fancy_grid'))
```

```python

missing_values = df.isna().sum()

print("Number of missing values in each column:")
print(missing_values)

total_missing = missing_values.sum()
print("\nTotal number of missing values in the DataFrame:", total_missing)
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import pandas as pd

file_path = 'preprocessed_reviews.csv'
df = pd.read_csv(file_path)

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

model = GPT2LMHeadModel.from_pretrained("gpt2", pad_token_id=tokenizer.eos_token_id)

from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.25, random_state=42)

train_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)
```

```python

total_samples = len(df)
train_samples = len(train_df)
test_samples = len(test_df)

train_ratio = train_samples / total_samples
test_ratio = test_samples / total_samples

print("Training data proportion:", train_ratio)
print("Testing data proportion:", test_ratio)
```

```python

sample_size = int(0.1 * len(df)) 

sampled_df = df.sample(n=sample_size, random_state=42)

train_df, test_df = train_test_split(sampled_df, test_size=0.25, random_state=42)

train_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)

total_samples = len(df)
train_samples = len(train_df)
test_samples = len(test_df)

train_ratio = train_samples / total_samples
test_ratio = test_samples / total_samples

print("Training data proportion:", train_ratio)
print("Testing data proportion:", test_ratio)
```
